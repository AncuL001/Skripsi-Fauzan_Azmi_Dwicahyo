```python
from tensorboard.backend.event_processing.event_accumulator import EventAccumulator
import matplotlib.pyplot as plt

def prep_tensorflow_log(path, to_plot=[]):
    event_acc = EventAccumulator(path)
    event_acc.Reload()

    for tag in to_plot:
        data = event_acc.Scalars(tag)
        x = [i.step for i in data]
        y = [i.value for i in data]
        plt.plot(x, y, label=tag)
```

```python
log_file = [
    "../runs/ResNet/dropout-0.1/2023-19-12_19-19-01/events.out.tfevents.1702988341.DESKTOP-FSI2RHS.21432.0"
]

prep_tensorflow_log(log_file[0], to_plot=["mAP/test"])

plt.xlabel("Epoch")
plt.ylabel("mAP")

plt.grid()
plt.legend(loc='best')

plt.show()
```
