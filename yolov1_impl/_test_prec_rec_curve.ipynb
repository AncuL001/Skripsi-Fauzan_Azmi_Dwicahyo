```python
import matplotlib.pyplot as plt
```

```python
precs = [1, 1, 0.5, 0.6667, 0.5, 0.4, 0.5, 0.4286, 0.375, 0.3333, 0.4]
recs = [0, 0.25, 0.25, 0.5, 0.5, 0.5, 0.75, 0.75, 0.75, 0.75, 1]

plt.plot(recs, precs)
plt.scatter(recs, precs)
plt.fill_between(recs, precs, alpha=0.3)
plt.grid()
axes = plt.gca()
axes.set_xlim([0,1])
axes.set_ylim([0,1])
axes.set_xlabel('Recall')
axes.set_ylabel('Precision')
plt.show()
```
